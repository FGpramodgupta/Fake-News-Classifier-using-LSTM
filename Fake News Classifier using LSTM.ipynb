```python
import pandas as pd
    
```

```python
df=pd.read_csv('fake-news/train.csv')
df.head()
```

```python
df=df.dropna()
```

```python
X=df.drop('label',axis=1)
X.shape
```

```python
y=df['label']
y.shape
```

```python
import tensorflow as tf

tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot

from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense

```

```python
voc_size=5000
```

 One Hot Representation

```python
message=X.copy()
message.head(100)
```

```python
message.reset_index(inplace=True)
message.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(message)):
    #print(i)
    
    if i % 200==0:
        print(i)
    review=re.sub('[^a-zA-Z]',' ',message['title'][i])
    review=review.lower()
    words=review.split()
    
    reviews=[ps.stem(word) for word in words if word not in stopwords.words('english')]
    review=' '.join(reviews)
    
    corpus.append(review)
    
```

```python
corpus
```

```python
one_hotRepr=[one_hot(word,voc_size)for word in corpus]
```

```python
one_hotRepr
```

```python
sent_length=20
embedded_docs=pad_sequences(one_hotRepr,padding='pre',maxlen=sent_length)
embedded_docs
#embedded_docs=pad_sequences(one_hotRepr,padding='post',maxlen=sent_legth)
```

```python
embedding_vectors_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vectors_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
#model.compile('adam','mse')
```

```python
model.summary()
```

```python
import numpy as np

X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.3,random_state=50)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```
